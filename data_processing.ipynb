### Filtering Volcanoes To Countries on Map

```python
volcanoes = df.read_csv('static/volcano.csv')

data = json.load(open('static/world-data.json'))

countries = []
for d in data['objects']['countries']['geometries']:
    countries.append(d['properties']['name'])

print(countries)

counts = volcanoes.value_counts('country').head(10)

missing_countries = []

filtered = volcanoes[~volcanoes.country.isin(countries)]
filtered.country.unique()


# popular_trees = list(counts.index)

# map = {}
# for i, tree in enumerate(popular_trees):
#     map[tree] = i

# with open("../static/species_indices.json", "w") as outfile: 
#     json.dump(map, outfile)

# popular_tree_species = trees[(trees.qSpecies.isin(popular_trees))]

# popular_tree_species.to_csv('../static/Popular_Trees.csv', index=False)
```

```python
volcanoes_filtered = volcanoes[volcanoes.country.isin(countries)]
print(volcanoes_filtered.shape)
volcanoes_filtered.head()
```

### Joining Eruptions data to Volcano Data

```python
#load eruptions
eruptions = df.read_csv('data/eruptions.csv')
eruptions.head()


cols_to_use = eruptions.columns.difference(volcanoes_filtered.columns)
print(cols_to_use)
volcanoes_erupted = df.merge(volcanoes_filtered, eruptions[cols_to_use], left_index=True, right_index=True, how='left')
volcanoes_erupted
```

```python
#drop columns
volcanoes_erupted = volcanoes_erupted.drop(columns = ["major_rock_2","major_rock_3","major_rock_4","major_rock_5", "minor_rock_1", "minor_rock_2", "minor_rock_3", "minor_rock_4", "minor_rock_5", "subregion", "eruption_number"])


volcanoes_erupted.columns
```

```python
volcanoes_erupted.head()
```

```python
#calculate eruption length

volcanoes_erupted = volcanoes_erupted.dropna(subset=['end_year']) #drop end years that are na 

volcanoes_erupted['start_date'] = df.to_datetime(volcanoes_erupted['start_year'].astype(int).astype(str) + '-' + volcanoes_erupted['start_month'].astype(int).astype(str) + '-' + volcanoes_erupted['start_day'].astype(int).astype(str))
volcanoes_erupted['end_date'] = df.to_datetime(volcanoes_erupted['end_year'].astype(int).astype(str) + '-' + volcanoes_erupted['end_month'].astype(int).astype(str) + '-' + volcanoes_erupted['end_day'].astype(int).astype(str))

volcanoes_erupted['eruption_length'] = (volcanoes_erupted['end_date'] - volcanoes_erupted['start_date']).dt.days

volcanoes_erupted.head()
```

```python
#create a field for number of erupions per volcano
print(len(volcanoes_erupted["volcano_name"].unique()))
print(len(eruptions["volcano_name"].unique()))
#more volcanoes in eruptions dataset than volcanoes_erupted

volcano_dict = {}
for i in volcanoes_erupted["volcano_name"]:
    volcano_dict[i] = 0
    
for i in volcano_dict:
    for j in eruptions["volcano_name"]:
        if i==j:
            volcano_dict[i]+=1
# print(volcano_dict)

print(sum(volcano_dict.values()))

volcanoes_erupted['num_eruptions'] = volcanoes_erupted['volcano_name'].map(volcano_dict)
volcanoes_erupted.head()
    
```

```python
#seeing volacano name counts between
print(eruptions["volcano_name"].value_counts())
# eruptions[eruptions["volcano_name"]=="Apo"].head()

print(volcanoes_filtered["volcano_name"].value_counts())
print(len(volcanoes_erupted[volcanoes_erupted["last_eruption_year"]=="Unknown"]))

# eruptions[eruptions["start_year"]=="Unkown"].head()
# volcanoes_filtered[volcanoes_filtered["volcano_name"]=="Apo"].head()
```

```python
#create a field for earliest known eruption year
early_year_dict={}
for i in volcanoes_erupted["volcano_name"]:
    early_year_dict[i] = "Unknown"

for i in early_year_dict:
    for j in eruptions["volcano_name"]:
        if i==j:
            temp = eruptions[eruptions["volcano_name"]==j]["start_year"]
            temp2 = volcanoes_erupted[volcanoes_erupted["volcano_name"]==j][["last_eruption_year"]]
            removed_unk = temp2[temp2["last_eruption_year"] != "Unknown"]
#             if "Unknown" in list(temp2["last_eruption_year"]):
#                 if len(removed_unk["last_eruption_year"])!=0:
#                     if min(temp) < min(removed_unk["last_eruption_year"].astype(float)):
#                         early_year_dict[j] = min(temp)
#                     else: 
#                         early_year_dict[j] = "Unknown"
            if len(removed_unk["last_eruption_year"])!=0:
                if min(temp) < min(removed_unk["last_eruption_year"].astype(int)):
                    early_year_dict[j] = int(min(temp))
                else:
                    early_year_dict[j] = int(min(removed_unk["last_eruption_year"].astype(int)))
 
print(early_year_dict)
zero_list=[]
for i in early_year_dict.items():
    if i[1] == "Unknown":
        zero_list.append(i)
        
print(len(zero_list))

```

```python
volcanoes_erupted['earliest_eruption_year'] = volcanoes_erupted['volcano_name'].map(early_year_dict)

volcanoes_erupted.head()
```

### Create Grouped Dataset for Spider Graph

```python
#calculate total number of volcanoes 
subset = volcanoes_erupted[["country", "eruption_length", "elevation", 
          "population_within_5_km",  
          "vei"]]

subset.head()
```

```python
#GROUP volcanoes by country and take average values
grouped_volcanoes = subset.groupby(by="country").mean().reset_index()
grouped_volcanoes.head()
```

```python
#calculate total number of volcanoes in each country
country_counts = []

for i in range(len(grouped_volcanoes["country"])):
    country_counts.append(0)
    for j in volcanoes_erupted["country"]:
        if grouped_volcanoes["country"][i]==j:
            country_counts[i]+=1
            
print(country_counts)
grouped_volcanoes["num_volcanoes"] = country_counts
grouped_volcanoes.head()
```

```python
#calculate number of eruptions in each country 
eruption_counts = []

for i in range(len(grouped_volcanoes["country"])):
    eruption_counts.append(0)
    for j in country_data["country"]:
        if grouped_volcanoes["country"][i]==j:
            eruption_counts[i]+=1
            
print(eruption_counts)
print(sum(eruption_counts))
print(grouped_volcanoes.shape)
grouped_volcanoes["num_eruptions"] = eruption_counts
grouped_volcanoes.head()

```
